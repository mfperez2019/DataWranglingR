```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from textblob import TextBlob
import nltk
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer 
```

```python
data = pd.read_csv("tweets4.csv",sep=',',encoding = "utf-8")
```

```python
data.head()
```

```python
#Función normalizadora
def word_norm(text):
  text = text.str.lower()
  text = text.replace('.', '')
  text = text.replace(',', '')
  text = text.replace('!', '')
  text = text.replace(to_replace =r'!', value='',regex=True)
  text = text.replace('á', 'a')
  text = text.replace('é', 'e')
  text = text.replace('í', 'i')
  text = text.replace('ó', 'o')
  text = text.replace('ú', 'u')
  text = text.replace('?', '')
  text = text.replace('!', '')
  text = text.replace(to_replace =r'-', value='',regex=True)
  text = text.replace(')', '')
  text = text.replace('(', '')
  text = text.replace(':', '')
  text = text.replace('/', '')
  return(text)
```

```python
data[['text']] = word_norm(data.text)
```

```python
data
```

```python
data['Source'] = data.statusSource.str.extract(r'(iphone|android)', expand=True)
```

```python
data
```

```python
#Generar un corpus del dataset original que contenga (10 pts):
#a. Handler
#b. screenName
#c. retweetCount
#d. Source
def corpus_creator(data):
    df= data
    out = df[['text','Handler','screenName','retweetCount','Source']]
    return out
```

```python
corpus = corpus_creator(data)
```

```python
corpus
```

```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from textblob import TextBlob
import nltk
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer 
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 

#Generar una función que lematize cada tweet y guardar el resultado en una nueva columna llamada "lem_text" (10 pts).
def tokenize(text):
    text1 = nltk.word_tokenize(text)
    text2 = ' '.join([lemmatizer.lemmatize(w) for w in text1])
    return text2
    
```

```python
#Columna que se llama lem_text
corpus['lem_text'] = corpus['text'].apply(tokenize)
```

```python
corpus
```

```python
#Generar una función calcule la polaridad y subjetividad de las columnas "text" y "lem_text" (20 pts).
def sentiment_analysis(text):
    polarity = []
    subjectivity = []
    for tw in text:
        tw = TextBlob(tw)
        ann = tw.sentiment
        polarity.append(ann[0])
        subjectivity.append(ann[1])

    return polarity,subjectivity
```

```python
corpus['Polarity'], corpus['Subjectivity']= sentiment_analysis(corpus['text'])
```

```python
corpus['Polarity1'], corpus['Subjectivity1']= sentiment_analysis(corpus['lem_text'])
```
